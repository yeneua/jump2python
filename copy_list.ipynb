# 복사

### 1) 리스트 복사

```python
a = [1, 2, 3]
b = a
```

b와 a는 동일한 객체를 가리킴<br/>
[1, 2, 3]이라는 리스트 객체를 참조하는 변수가 하나 더 늘어난 것<br/>
실제로 리스트는 1개만 존재

### 2) [:] 복사하기

```python
a = [1, 2, 3]
b = a[:]

a[0] = 5
```

`b = a[:]` 에서 새로 복사한 값을 b에 넣어줌<br/>
a와 b는 서로 다른 객체를 가리킴

### 3) 중첩 리스트 복사

```python
a = [1, 2, [3, 4]]
b = a[:]
```

`id(a)`와 `id(b)` 값은 다르지만<br/>
`id(a[2])`와 `id(b[2])`의 값은 같다 = 같은 객체를 가리킴

```python
b[2][0] = 999
```

같은 곳을 가리키고 있기 때문에 `a[2][0]`, `b[2][0]`의 값이 999로 같다

```python
b[2] = [9, 9]
```

`b[2]`가 새로운 주소를 가지는 객체로 할당된다

```python
b[0] = 99
```

`b[0]` 값만 바뀐다

### 정리
list 슬라이싱으로 새로운 값을 할당해줌<br/>
새로운 id 값이 부여되면서 서로 영향을 받지 않음<br/>
하지만 mutable 객체 안에 mutable인 경우 문제 발생<br/>
=> `id(a) != id(b)`, `id(a[0]) == id(b[0])`
