### 3. 객체에 연산할 숫자 지정하기

사칙 연산을 할 때 사용할 2개의 숫자를 a객체에 알려줘야 함<br/>
연산 수행 대상 객체에 지정할 수 있게 하기

```python
class FourCal:
    def setdata(self, first, second): # setdata 메서드의 매개변수
        self.first = first # setdata 메서드의 수행문
        self.second = second # setdata 메서드의 수행문
```

```python
# a객체 생성 및 메서드 호출
a = FourCal()
a.setdata(4, 2)
```

`self`에는 `setdata` 메서드를 호출한 `객체 a`가 자동으로 전달됨

```python
a = FourCal()
FourCal.setdata(a, 1, 2) # 클래스명.메서드 형태로 호출 시 -> 객체를 self 매개변수에 꼭 전달
```

**set 메서드의 수행문**<br/>
first, second에 값이 저장됨<br/>
=> 객체에 생성되는 객체만의 변수를 `객체변수` 또는 `속성`이라고 함

```python
a = FourCal()
a.setdata(4,2)
print(a.first)
print(a.second)
```

```python
b = FourCal()
b.setdata(3,7)
print(b.first)
print(b.second)
```

클래스로 만든 객체의 객체변수는 다른 객체의 객체변수와 상관없이 독립적인 값 유지

### 4. 더하기 기능 만들기

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
```

```python
a = FourCal()
a.setdata(4, 2)
a.add()
```

### 5. 곱하기, 빼기, 나누기 기능 만들기

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
a.setdata(4,2)
print(a.add())
print(a.sub())
print(a.mul())
print(a.div())
```

## 생성자

생성자(constructor) : 객체가 생성될 때 자동으로 호출되는 메서드

```python
a = FourCal()
a.add()
```

a에 setdata 메서드를 수행하지 않아서 오류 발생<br/>
초깃값 설정해야 할 필요가 있는 경우 -> 메서드 호출하여 설정하기보다 생성자를 구현하는 것이 안전한 방법

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal() # first, second 값 전달되지 않아 오류 발생
```

```python
a = FourCal(3,5)
print(a.first, a.second)
```

## 클래스의 상속

`class 클래스이름(부모클래스이름):`

```python
class MoreFourCal(FourCal):
    pass
```

```python
c = MoreFourCal(4,2)
print(c.add())
print(c.sub())
print(c.mul())
print(c.div())
```

```python
class MoreFourCal(FourCal):
    def pow(self): ## 메서드 추가
        result = self.first ** self.second
        return result
```

```python
c = MoreFourCal(4,2)
c.pow()
```

## 메서드 오버라이딩(method overriding)

부모 클래스에 있는 메서드를 동일한 이름으로 다시 만드는 것<br/>
부모 클래스 메서드 대신 오버라이딩한 메서드 호출됨

```python
a = FourCal(4,0)
a.div()
```

```python
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0: # second가 0이면 0을 리턴하도록 오버리아딩
            return 0
        else:
            return self.first / self.second
```

```python
a = SafeFourCal(4,0)
a.div()
```

## 클래스 변수

`클래스이름.클래스변수`
<br/>클래스 변수 : 모든 객체에 공유됨<br/>
객체 변수 : 다른 객체들의 영향을 받지 않고 독립적으로 그 값 유지

```python
class Family:
    lastname='김'
```

```python
Family.lastname
```

```python
a = Family()
b = Family()
print(a.lastname)
print(b.lastname)
```

클래스 변수 값 변경 -> 같은 클래스로 만든 객체 lastname 값 모두 변경됨

```python
Family.lastname = '박'
print(a.lastname)
print(b.lastname)
```

클래스 변수와 동일한 이름의 객체변수?

```python
a.lastname = '최'
a.lastname
```

```python
print(Family.lastname)
print(b.lastname)
```

a 객체의 lastname이라는 객체 변수가 새롭게 생성됨<br/>
Family 클래스의 lastname값은 변경x
